```python
import json
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

with open("human_detection.json", "r") as f:
    human_dections = json.load(f)


def draw_bboxes(bboxes, image):
    for bbox in bboxes:
        image = cv2.rectangle(image, (bbox[0], bbox[1]), (bbox[2], bbox[3]), (255, 0, 0), 2)
    return image


def change_bbox_format(bboxes):
    changed_bboxes = []
    for bbox in bboxes:
        x1 = int(bbox[0])
        y1 = int(bbox[1])
        x2 = int(x1 + bbox[2])
        y2 = int(y1 + bbox[3])
        changed_bboxes.append([x1, y1, x2, y2])
    return changed_bboxes


def adjust_bbox(bboxes):
    # bboxes = change_bbox_format(bboxes)
    bboxes = np.asarray(bboxes)
    x1 = bboxes[:, 0]
    y1 = bboxes[:, 1]
    x2 = bboxes[:, 2]
    y2 = bboxes[:, 3]
    new_bbox = [min(x1), min(y1), max(x2), max(y2)]
    return new_bbox


def cal_avg_bbox(bboxes):
    for i in range(15, len(bboxes) - 15):
        sub_bboxes = bboxes[i-15:i+15]
        if i == 145:
            return sub_bboxes


def collect_bbox(human_detections):
    bboxes = []
    for detection in human_dections:
        bboxes.append(detection["bbox"])
    return bboxes


bboxes = collect_bbox(human_detections=human_dections)
bboxes = change_bbox_format(bboxes=bboxes)
sub_bboxes = cal_avg_bbox(bboxes=bboxes)
print((sub_bboxes))

img = cv2.cvtColor(cv2.imread("videoframes/frame000145.jpg"), cv2.COLOR_BGR2RGB)
img_rec = draw_bboxes(bboxes=sub_bboxes, image=img.copy())
new_bbox = adjust_bbox(bboxes=sub_bboxes)
print(new_bbox)
img_rec = cv2.rectangle(img_rec, (new_bbox[0], new_bbox[1]), (new_bbox[2], new_bbox[3]), (0, 255, 0), 3)
plt.imshow(img_rec)
```
